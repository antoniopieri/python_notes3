# Plotting Multiple Variables <h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('bad-drivers.csv').head(3)

#print(df.head())
```

```python
sns.regplot(data=df,
            x='Losses incurred by insurance companies for collisions per insured driver ($)',
            y='Car Insurance Premiums ($)')

```

```python
sns.lmplot(data=df,
           x='Losses incurred by insurance companies for collisions per insured driver ($)',
           y='Car Insurance Premiums ($)',
           hue="State")
plt.show()

#lmplot is similar to a regplot
#lmplot is able to allow us to 'hue' multiple graphs on top of each other [stack]
```

```python
sns.lmplot(data=df,
           x='Losses incurred by insurance companies for collisions per insured driver ($)',
           y='Car Insurance Premiums ($)',
           row="State")
```

```python
sns.lmplot(data=df,
           x='Losses incurred by insurance companies for collisions per insured driver ($)',
           y='Car Insurance Premiums ($)',
           col="State")
```
