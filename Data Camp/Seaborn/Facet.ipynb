```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train=pd.read_csv('train.csv')

```

# Facet Plots

Using FacetGrid to plot multiple pointplots

```python
g=sns.FacetGrid(data=train, col='Embarked',col_order=['C','S','Q'])
g.map(sns.pointplot,'Pclass' ,'Survived','Sex', hue_order=None)
g.add_legend(title='Gender')
```

# Cat Plot (factplot depreciated) 

```python
sns.catplot(data=train,
         x='Age',
         kind='box',
         col='Pclass')
train.columns
```

```python
sns.catplot(data=train,
         x='Pclass', y='Age',
         kind='bar',hue='Sex',
         col='Embarked')
```

```python
g=sns.FacetGrid(data=train,col='Pclass')
g.map(plt.scatter,x='Age',y='Fare')

#why dosnt this work!?!?
```

```python
sns.lmplot('Age','Fare',data=train, col='Pclass')
```

```python
sns.lmplot(data=train,x='Age',y='Fare',col='Pclass',hue='Sex')
```
