# Intro to Seaborn <h1>

Distribution plots of SINGLE variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('bad-drivers.csv')
df['Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding'].plot.hist() 
#pandas has a histogram function
#plt.clf() this clear the figure
```

```python
sns.distplot(df['Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding'])

#the above plots a distribution plot using seaborn
#the default includes both a smooth line and a histogram
```

```python
sns.distplot(df['Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding'],
            kde=False, #Kernel Density Estimate(KDE) creates a smoothened line. Here we remove it 
            bins=20) #here we specify the bins (number ofsections in histogram)
```

```python
sns.distplot(df['Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding'],
            hist=False,
            rug=True,
            kde_kws={'shade':True})
```
